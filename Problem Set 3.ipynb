### Occupations

1. **Step 1:** Import the necessary libraries.
2. **Step 2:** Import the dataset from [this address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user).
3. **Step 3:** Assign it to a variable called `users`.
4. **Step 4:** Discover what is the **mean age per occupation**.
5. **Step 5:** Discover the **Male ratio per occupation** and sort it from the most to the least.
6. **Step 6:** For each occupation, calculate the **minimum and maximum ages**.
7. **Step 7:** For each combination of occupation and sex, calculate the **mean age**.
8. **Step 8:** For each occupation, present the **percentage of woen and men**.

```python
# Step 1: Import the necessary libraries
import pandas as pd

# Step 2: Load the dataset from the specified URL
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"

# Step 3: Read the dataset into a DataFrame and assign it to users
users = pd.read_csv(url, sep="|")

# Display the initial rows to inspect the dataset structure
print("Preview of the dataset:")
print(users.head(), "\n")

# Step 4: Calculate the mean age per occupation
mean_age_per_occupation = users.groupby('occupation')['age'].mean()
print("Mean age per occupation:")
print(mean_age_per_occupation, "\n")

# Step 5: Discover the Male ratio per occupation and sort it from the most to the least
users['is_male'] = users['gender'].apply(lambda x: 1 if x == 'M' else 0)
male_ratio_per_occupation = users.groupby('occupation')['is_male'].mean()
sorted_male_ratio = male_ratio_per_occupation.sort_values(ascending=False)
print("Male ratio per occupation (sorted):")
print(sorted_male_ratio, "\n")

# Step 6: Calculate the minimum and maximum ages per occupation
min_max_age_per_occupation = users.groupby('occupation')['age'].agg(['min', 'max'])
print("Minimum and maximum ages per occupation:")
print(min_max_age_per_occupation, "\n")

# Step 7: Calculate the mean age for each combination of occupation and sex
mean_age_per_occ_sex = users.groupby(['occupation', 'gender'])['age'].mean()
print("Mean age per occupation and gender:")
print(mean_age_per_occ_sex, "\n")

# Step 8: Calculate the percentage of women and men for each occupation
gender_counts = users.groupby(['occupation', 'gender']).size().unstack(fill_value=0)
gender_percentage = gender_counts.div(gender_counts.sum(axis=1), axis=0) * 100
print("Percentage of women and men for each occupation:")
print(gender_percentage)

```

   Calculate the weekly minimum, maximum, mean, and standard deviations for the windspeeds across all locations for the first 52 weeks of data.


```python
import pandas as pd

# Step 1: Import the necessary libraries
file_path = r"C:\Users\rinci\Downloads\wind.txt"
wind_data = pd.read_csv(file_path, sep=r'\s+', header=0)

# Step 2: Assign the dataset to a variable and replace the first 3 columns by a proper datetime index
wind_data.columns = ['Year', 'Month', 'Day'] + list(wind_data.columns[3:])

# Step 3: Inspect and fix the 'Year', 'Month', and 'Day' columns
# Correct 'Year' by adding 1900 to values less than 100
wind_data['Year'] = wind_data['Year'].apply(lambda y: y + 1900 if y < 100 else y)

# Ensure 'Month' and 'Day' are numeric and valid
wind_data['Month'] = pd.to_numeric(wind_data['Month'], errors='coerce')  # Convert to numeric, NaN for invalid values
wind_data['Day'] = pd.to_numeric(wind_data['Day'], errors='coerce')

# Drop rows with missing or invalid 'Month' or 'Day' values
wind_data_cleaned = wind_data.dropna(subset=['Month', 'Day'])

# Convert 'Month' and 'Day' to integers
wind_data_cleaned['Month'] = wind_data_cleaned['Month'].astype(int)
wind_data_cleaned['Day'] = wind_data_cleaned['Day'].astype(int)

# Step 4: Create the 'Date' column with proper mapping
try:
    wind_data_cleaned['Date'] = pd.to_datetime(wind_data_cleaned[['Year', 'Month', 'Day']].rename(columns={'Year': 'year', 'Month': 'month', 'Day': 'day'}), errors='coerce')
    print("Successfully created the 'Date' column.")
except Exception as e:
    print(f"Error creating 'Date' column: {e}")

# Step 5: Drop rows with invalid dates
wind_data_final = wind_data_cleaned.dropna(subset=['Date'])

# Step 6: Set 'Date' as the index
wind_data_final = wind_data_final.set_index('Date').drop(columns=['Year', 'Month', 'Day'])

# Step 7: Compute how many values are missing for each location over the entire record
missing_windspeeds = wind_data_final.isnull().sum()
print(f"Missing values per location:\n{missing_windspeeds}")

# Step 8: Compute how many non-missing values there are in total
total_non_missing = wind_data_final.notna().sum().sum()
print(f"Total non-missing values: {total_non_missing}")

# Step 9: Calculate the mean windspeeds over all the locations and all the times
overall_mean_windspeed = wind_data_final.mean().mean()
print(f"Mean windspeed across all locations and times: {overall_mean_windspeed:.2f}")

# Step 10: Create a DataFrame called loc_stats and calculate the min, max, mean windspeeds, and standard deviations at each location
location_statistics = wind_data_final.describe().T[['min', 'max', 'mean', 'std']]
print("\nLocation statistics:\n", location_statistics)

# Step 11: Create a DataFrame called day_stats and calculate the min, max, mean windspeeds, and standard deviations across all locations at each day
daily_statistics = wind_data_final.agg(['min', 'max', 'mean', 'std'], axis=1)
print("\nDaily statistics (first 5 rows):\n", daily_statistics.head())

# Step 12: Find the average windspeed in January for each location
january_wind_data = wind_data_final[wind_data_final.index.month == 1]
january_average_wind = january_wind_data.mean()
print("\nJanuary average windspeed per location:\n", january_average_wind)

# Step 13: Downsample the record to a yearly frequency for each location
yearly_wind_data = wind_data_final.resample('YE').mean()
print("\nYearly average windspeed:\n", yearly_wind_data.head())

# Step 14: Downsample the record to a monthly frequency for each location
monthly_wind_data = wind_data_final.resample('ME').mean()
print("\nMonthly average windspeed:\n", monthly_wind_data.head())

# Step 15: Downsample the record to a weekly frequency and calculate the min, max, mean, and standard deviations for the first 52 weeks starting on January 2, 1961
start_date = pd.Timestamp('1961-01-02')  # First week starts on January 2, 1961
end_date = start_date + pd.DateOffset(weeks=52)  # Calculate the end date (52 weeks later)

# Filter data for the first 52 weeks
filtered_data = wind_data_final.loc[start_date:end_date]

# Resample the data weekly, with weeks starting on January 2, 1961
weekly_stats = filtered_data.resample('W-MON').agg(['min', 'max', 'mean', 'std'])

# Print the weekly statistics for the first 52 weeks
print("\nWeekly statistics for the first 52 weeks (starting January 2, 1961):")
print(weekly_stats.head(52))  # Print the first 52 weeks
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Import the dataset 
dataset_path = r"C:\Users\rinci\Downloads\arcade-revenue-vs-cs-doctorates.csv"
arcade_data = pd.read_csv(dataset_path)

# Step 2: Extract data specifically for the years 2000 to 2009
filtered_data = arcade_data[(arcade_data['Year'] >= 2000) & (arcade_data['Year'] <= 2009)]

# Step 3: Plot a scatter plot showing arcade revenue against CS doctorates awarded
plt.figure(figsize=(10, 6))
scatter_chart = plt.scatter(
    filtered_data['Total Arcade Revenue (billions)'],
    filtered_data['Computer Science Doctorates Awarded (US)'],
    c=filtered_data['Year'], 
    cmap='plasma',  
    edgecolor='black', 
    s=100
)

# Step 4: Annotate the chart with a descriptive title and label the axes
plt.title('Arcade Revenue vs. Computer Science Doctorates (2000-2009)', fontsize=14)
plt.xlabel('Arcade Revenue (in billions)', fontsize=12)
plt.ylabel('CS Doctorates Awarded (US)', fontsize=12)

# Step 5: Add a color bar to represent the years included in the data
plt.colorbar(scatter_chart, label='Year (2000-2009)')

# Step 6: Enhance visibility with gridlines and layout adjustments
plt.grid(True)
plt.tight_layout()

# Step 7: Render the final scatter plot
plt.show()
```
